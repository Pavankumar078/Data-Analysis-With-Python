##### Analysing the Gender

```python
import matplotlib.pyplot as plt
```

```python
#Ratio of Gender male:female
data=pd.DataFrame({"Ratio":[len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])]},index=['Male','Female'])
print(data)
data.plot.pie(y='Ratio',figsize=(7,7),autopct="%.2f")
```

```python
# Other way
df.groupby('Gender').size()
```

```python
df.groupby('Gender').size().plot(kind='pie',autopct='%.2f',
                                title='Gender Ratio',
                                figsize=(6,6))
```

```python
# Barchart
df.groupby('Gender').size().plot(kind='bar',figsize=(6,6))
```

```python
#Total puchases by gender
df.groupby('Gender').sum()['Purchase']
```

```python
#Plotting the pie char for total puchases by gender
df.groupby('Gender').sum()['Purchase'].plot(kind='pie',autopct='%.2f')
```

```python
#Mean of the purchases by gender
df.groupby('Gender').mean()['Purchase'].plot(kind='pie',autopct="%0.2f",figsize=(6,6))
```

##### Analysing the age marital status

```python
#Total number of persons in each age groups
df.groupby('Age').size()
```

```python
# Puchase Distribution per age
df.groupby('Age').size().plot(kind='bar',figsize=(12,6),title="Purchase Distribution by age")
```

```python
#Extracting the Total Purchases per age group and storing it in other dataframe
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst, columns=['Age', 'Products'])
data
```

```python
data.plot.bar(x='Age',figsize=(8,6))
```

```python
# Total amount spend by each age group
df.groupby('Age').sum()['Purchase']
```

```python
#Bar chart for the total amount spend by each age group in purchasing
df.groupby('Age').sum()['Purchase'].plot(kind='bar',figsize=(12,6),title="Amount spend by age")
```

```python
#Mean of the purchases by age group
df.groupby('Age').mean()['Purchase']
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind='pie',autopct='%.2f',figsize=(6,6))
```
